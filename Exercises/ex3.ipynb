```python
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'

!pip install python-bidi

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import widgets, VBox, Output
from IPython.display import display
from bidi.algorithm import get_display

pd.set_option("display.max_columns", None)

response = requests.get(url)
data = response.json()
data_df = pd.DataFrame(data['result']['records'])

# Tab 1: Data Overview
tab1_content = widgets.Output()
with tab1_content:
    print("Data Overview:")
    display(data_df.describe())  # Summary statistics

# Tab 2: Raw Data
tab2_content = widgets.Output()
with tab2_content:
    print("Raw Data:")
    display(data_df)  # Full DataFrame

# Tab 3: Charts
tab3_content = widgets.Output()
with tab3_content:
    print("Top 10 Car Manufacturers (Fixed Bidi):")
    chart_data = data_df['tozeret_nm'].value_counts().head(10)

    # Apply bidi to the index (Manufacturer names) and keep the values (counts)
    fixed_index = [get_display(str(label)) for label in chart_data.index]
    fixed_series = pd.Series(chart_data.values, index=fixed_index)

    fig, ax = plt.subplots(figsize=(10, 6))
    fixed_series.plot(kind='bar', ax=ax)
    ax.set_title(get_display('עשרת יצרני הרכב המובילים')) # Example of bidi in title
    ax.set_xlabel(get_display('יצרן'))
    ax.set_ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content, tab3_content])
tabs.set_title(0, 'Tab 1: Data Overview')
tabs.set_title(1, 'Tab 2: Raw Data')
tabs.set_title(2, 'Tab 3: Charts')

# Display Tabs
display(tabs)
```
