```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import os

# Path to the students.txt file in Google Drive
file_path = '/content/drive/My Drive/students.txt'

# Data Preparation
def load_student_data(self):
        """Reads data from the JSON file safely."""
        self.student_details_map.clear()
        self.display_names.clear()

        if os.path.exists(self.file_path):
            try:
                with open(self.file_path, 'r') as f:
                    # Load the JSON dictionary
                    data = json.load(f)

                    # Create a unique key for the dropdown (e.g., "John Doe")
                    display_name = f"{data.get('first_name')} {data.get('last_name')}"

                    # Store the whole dictionary mapped to that name
                    self.student_details_map[display_name] = data
                    self.display_names = [display_name]
            except Exception as e:
                print(f"Error loading JSON: {e}")

# --- Styling ---
form_style = """
<style>
    .student-form-container {
        font-family: 'Segoe UI', Arial, sans-serif;
        background-color: #f9f9f9;
        border: 1px solid #ddd;
        border-radius: 12px;
        padding: 30px;
        max-width: 50%;
        margin: 20px auto;
        box-shadow: 0 4px 20px rgba(0,0,0,0.05);
    }
    .info-card {
        background-color: #ffffff;
        border-radius: 8px;
        padding: 20px;
        border-left: 5px solid #007bff;
        margin-top: 15px;
    }
</style>
"""

if not display_names:
    display(HTML("<div style='color: red;'>Error: No valid student data found.</div>"))
else:
    student_selection_dropdown = widgets.Dropdown(
        options=display_names,
        description='Select Student:',
        style={'description_width': 'initial'},
        layout={'width': '50%'}
    )

    fav_show_input = widgets.Text(
        value='',
        placeholder='Type something',
        description='Fav Show:',
        disabled=False,
        layout={'width': '70%'}
    )

    save_button = widgets.Button(
        description='Save',
        button_style='success',
        layout={'width': '25%'}
    )

    student_info_output = widgets.Output()
    save_status_output = widgets.Output()

    def update_display(self, change):
            """Updates the info card using JSON dictionary keys."""
            selected = change.get('new')
            details = self.student_details_map.get(selected)

            with self.info_output:
                clear_output(wait=True)
                if details:
                    # Direct access to JSON keys
                    email = details.get('email', 'N/A')
                    courses = ", ".join(details.get('courses_this_semester', []))
                    link = details.get('interesting_link', '#')

                    html = f"""
                    <div class='info-card'>
                        <div style='margin-bottom:8px;'><strong>Name:</strong> {selected}</div>
                        <div style='margin-bottom:8px;'><strong>Email:</strong> {email}</div>
                        <div style='margin-bottom:8px;'><strong>Courses:</strong> {courses}</div>
                        <div style='margin-bottom:8px;'><strong>Link:</strong> <a href="{link}" target="_blank">{link}</a></div>
                    </div>
                    """
                    display(widgets.HTML(html))

    def save_favorite_show(b):
        selected_student = student_selection_dropdown.value
        show_name = fav_show_input.value.strip()
        if show_name and selected_student:
            with save_status_output:
                clear_output()
                try:
                    with open(file_path, 'a') as f:
                        f.write(f"\n{selected_student} Favourite Show: {show_name}")
                    print(f"Saved '{show_name}' for {selected_student}!")
                except Exception as e:
                    print(f"Error saving: {e}")

    student_selection_dropdown.observe(update_display, names='value')
    save_button.on_click(save_favorite_show)

    # Assemble the UI elements into a VBox
    form_content = widgets.VBox([
        student_selection_dropdown,
        student_info_output,
        widgets.HBox([fav_show_input, save_button]),
        save_status_output
    ])

    # Assemble the UI with styling
    display(HTML(form_style))
    display(HTML("<div class='student-form-container'><h2>Student Directory</h2>"))
    display(form_content)
    display(HTML("</div>"))

    # Trigger initial view
    class Init:
        def __init__(self, val): self.new = val
    update_display(Init(student_selection_dropdown.value))
```
